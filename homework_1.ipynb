```python
import pandas as pd
from IPython.display import display
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/HalyshAnton/IT-Step-Pyton-AI/main/module2/data/sales_data.csv", index_col="Unnamed: 0")
```

#Завдання **1**

```python
print("Загальна інформація про таблицю:")
print(df.info())
print(f"Кількість рядків: {df.shape[0]}")
print(f"Кількість стовпців: {df.shape[1]}")
print("Назви стовпців:", df.columns.tolist())
print("Стовпці з пропущеними значеннями:", df.columns[df.isnull().any()].tolist())
print("\n" + "-"*50 + "\n")
```

## Завдання 2

```python
df.columns = df.columns.str.lower()
print("Назви стовпців після зниження регістру:", df.columns.tolist())
print("\n" + "-"*50 + "\n")
```

## Завдання 3

```python
tel_โทร = ['телевізор', 'телефон']
filtered_data = df[df['item'].isin(tel_โทร)]
print("Купівлі телевізорів та телефонів:")
display(filtered_data)
print("\n" + "-"*50 + "\n")
```

## Завдання 4

```python
luis_sales = df[df['manager'] == 'Luis']
total_revenue_luis = luis_sales['sale_amt'].sum()
print(f"Загальна виручка продавця Luis: {total_revenue_luis}")
print("\n" + "-"*50 + "\n")
```

## Завдання 5

```python
grouped_items = df.groupby('item')['units'].sum().reset_index().sort_values(by='units', ascending=False)
print("Кількість проданих товарів та їх типи:")
display(grouped_items)
print("\n" + "-"*50 + "\n")
```

## Завдання 6

```python
df['calculated_sale_amt'] = df['units'] * df['unit_price']
errors = df[df['sale_amt'] != df['calculated_sale_amt']]
print("Рядки з помилками у розрахунках:")
display(errors)
```

## Виправлення помилок у sale_amt

```python
df.loc[errors.index, 'sale_amt'] = df.loc[errors.index, 'calculated_sale_amt']
print("Помилки виправлені. Перевірка ще раз:")
errors_post = df[df['sale_amt'] != df['calculated_sale_amt']]
if errors_post.empty:
    print("Помилок більше немає.")
else:
    print(errors_post)
```
