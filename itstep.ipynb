```python
import torch
```

```python
def f(x):
    return x**2 / 3 + torch.sin(torch.pi * x)

y_true = 1
```

```python
start_point = torch.tensor(1.2, device='cuda', requires_grad=True)
```

```python
optimizer = torch.optim.SGD([start_point], lr=1e-3)
```

```python
for i in range(2000):
    res_func = f(start_point)

    error = (y_true - res_func) ** 2

    error.backward()

    optimizer.step()

    optimizer.zero_grad()

    if i % 100 == 0:
        print(f'Ітерація {i}, помилка: {error.item():.6f}')

print()
print(f'Оптимальне x: {start_point.item():.6f}')
print(f'f(x): {f(start_point).item():.6f}')
```
